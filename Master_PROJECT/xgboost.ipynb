```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, roc_curve, roc_auc_score

import kagglehub
```

```python
path = kagglehub.dataset_download("janus137/supersymmetry-dataset")

print("Path to dataset files:", path)
```

```python
filename="supersymmetry_dataset.csv"
```

```python
features = ['SUSY','lepton 1 pT', 'lepton 1 eta', 'lepton 1 phi', 'lepton 2 pT', 'lepton 2 eta', 'lepton 2 phi',
            'missing energy magnitude', 'missing energy phi', 'MET_rel', 'axial MET', 'M_R', 'M_TR2', 'R',
            'MT2', 'S_R', 'M_Delta_R', 'dPhi_r_b', 'cos(theta_r1)']

n_examples =1000000
df = pd.read_csv("/root/.cache/kagglehub/datasets/janus137/supersymmetry-dataset/versions/1/supersymmetry_dataset.csv", header=None, names=features, nrows=n_examples, skiprows=1)
df
```

```python
X = df.drop('SUSY', axis=1)
y = df['SUSY'].astype(int)  # target explicit int
low_features = ["lepton 1 pT", "lepton 1 eta", "lepton 1 phi", "lepton 2 pT", "lepton 2 eta", "lepton 2 phi",
                "missing energy magnitude","missing energy phi"]
high_features = ['MET_rel', 'axial MET', 'M_R', 'M_TR2', 'R', 'MT2','S_R', 'M_Delta_R', 'dPhi_r_b', 'cos(theta_r1)']

X_low  = X[low_features]
X_high = X[high_features]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.1, random_state=42)

X_low_train, X_low_test, _, _  = train_test_split(X_low, y, stratify=y, test_size=0.1, random_state=42)

X_high_train, X_high_test, _,_ = train_test_split(X_high, y, stratify=y, test_size=0.1, random_state=42)
```

```python
import xgboost as xgb
from sklearn.metrics import accuracy_score, roc_auc_score
import time
```

```python
clf_full = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=1, n_jobs=-1)
start = time.time()
clf_full.fit(X_train, y_train)
runtime_full = time.time() - start
y_pred_full = clf_full.predict(X_test)
print(f"Full features | Accuracy: {100*accuracy_score(y_test, y_pred_full):.2f}% | AUC: {roc_auc_score(y_test, y_pred_full):.2f} | Time: {runtime_full:.2f}s")

clf_low = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=1, n_jobs=-1)
start = time.time()
clf_low.fit(X_low_train, y_train[:len(X_low_train)])
runtime_low = time.time() - start
y_pred_low = clf_low.predict(X_low_test)
print(f"Low-level features | Accuracy: {100*accuracy_score(y_test[:len(X_low_test)], y_pred_low):.2f}% | AUC: {roc_auc_score(y_test[:len(X_low_test)], y_pred_low):.2f} | Time: {runtime_low:.2f}s")

clf_high = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=1, n_jobs=-1)
start = time.time()
clf_high.fit(X_high_train, y_train[:len(X_high_train)])
runtime_high = time.time() - start
y_pred_high = clf_high.predict(X_high_test)
print(f"High-level features | Accuracy: {100*accuracy_score(y_test[:len(X_high_test)], y_pred_high):.2f}% | AUC: {roc_auc_score(y_test[:len(X_high_test)], y_pred_high):.2f} | Time: {runtime_high:.2f}s")
```

```python
plt.figure(figsize=(7,4))
xgb.plot_importance(clf_full)
plt.title("Feature Importance (All Features)")
plt.show()

plt.figure(figsize=(7,4))
xgb.plot_importance(clf_low)
plt.title("Feature Importance (Low-level)")
plt.show()

plt.figure(figsize=(7,4))
xgb.plot_importance(clf_high)
plt.title("Feature Importance (High-level)")
plt.show()
```

```python
from sklearn.metrics import roc_curve

prob_full = clf_full.predict_proba(X_test)[:,1]
prob_low  = clf_low.predict_proba(X_low_test)[:,1]
prob_high = clf_high.predict_proba(X_high_test)[:,1]
fpr_full, tpr_full, _ = roc_curve(y_test, prob_full)
fpr_low, tpr_low, _   = roc_curve(y_test[:len(X_low_test)], prob_low)
fpr_high, tpr_high, _ = roc_curve(y_test[:len(X_high_test)], prob_high)

plt.figure(figsize=(7,5))
plt.plot(tpr_full, 1-fpr_full, label="Full")
plt.plot(tpr_low, 1-fpr_low, label="Low")
plt.plot(tpr_high, 1-fpr_high, label="High")
plt.xlabel("Signal efficiency (TPR)")
plt.ylabel("Background rejection (1-FPR)")
plt.legend()
plt.title("XGBoost ROC curves on SUSY (subset)")
plt.grid(True, alpha=0.4)
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV

params = {
    'max_depth':[3,4,6],
    'min_child_weight':[1,3,5],
    'learning_rate':[0.1,0.3]
}
clf_grid = xgb.XGBClassifier(objective='binary:logistic', eval_metric='logloss', n_estimators=100,
                            colsample_bytree=1, random_state=1, n_jobs=-1, use_label_encoder=False)

grid = GridSearchCV(clf_grid, params, cv=5, scoring="accuracy", verbose=2, n_jobs=-1)
grid.fit(X_train, y_train)
print("Best training set score: {:.3f}".format(grid.best_score_))
print("Best parameters: ", grid.best_params_)
print("Test set accuracy: {:.2f}%".format(100*grid.best_estimator_.score(X_test, y_test)))
print("Test set AUC: {:.2f}".format(roc_auc_score(y_test, grid.best_estimator_.predict(X_test))))

from sklearn.metrics import roc_curve

prob_full = clf_full.predict_proba(X_test)[:,1]
prob_low  = clf_low.predict_proba(X_low_test)[:,1]
prob_grid = grid.best_estimator_.predict_proba(X_test)[:, 1]

fpr_full, tpr_full, _ = roc_curve(y_test, prob_full)
fpr_low, tpr_low, _   = roc_curve(y_test, prob_low)
fpr_grid, tpr_grid, _ = roc_curve(y_test, prob_grid)


plt.figure(figsize=(8,6))
plt.plot(tpr_full, 1-fpr_full, label="Full (Varsayılan)")
plt.plot(tpr_low, 1-fpr_low, label="Low-level")
plt.plot(tpr_grid, 1-fpr_grid, label="GridSearch Best Model", linestyle='--', linewidth=2, color='red')

plt.xlabel("Signal efficiency (TPR)")
plt.ylabel("Background rejection (1-FPR)")
plt.legend()
plt.title("XGBoost ROC Curves on SUSY Dataset")
plt.grid(True, alpha=0.4)
plt.show()

clf_es = xgb.XGBClassifier(n_estimators=2000, early_stopping_rounds=10, eval_metric="logloss", use_label_encoder=False)
clf_es.fit(X_train, y_train, eval_set=[(X_test, y_test)], verbose=True)

y_pred_es = clf_es.predict(X_test)
print("\n--- Early Stopping Model Sonuçları ---")
print(f"Accuracy: {100*accuracy_score(y_test, y_pred_es):.2f}% | AUC: {roc_auc_score(y_test, y_pred_es):.2f}")

```
